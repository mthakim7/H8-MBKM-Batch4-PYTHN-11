# Pandas

Pandas adalah tools analisis dan manipulasi data open source yang cepat, powerfull, fleksibel, dan mudah digunakan, dibangun di atas bahasa pemrograman Python.

Pandas adalah pustaka Python yang menyediakan saran ekstensif untuk analisis data. Data scientist sering kali bekerja dengan data yang disimpan dalam format tabel seperti .csv .tsv dan .xlsx . Pandas membuatnya sangat nyaman untuk load, memproses, dan menganlisis data tabel tersebut menggunakan kueri seperti SQL. Dalam hubungannya dengan Matplotlib dan Seaborn, Pandas memberikan berbagai peluang untuk analisis visual data tabel.

Struktur data utama di Pandas diimplementasikan dengan Series dan DataFrame. DataFrames sangat bagus untuk mempresentasikan data nyata: baris seusai dengan instance (objek, observasi, dll) dan kolom sesuai dengan features untuk setiap instance

Pandas seperti Excel pada Python: dapat menggunanakan tabel (yang dinamakan DataFrame) dan operator transformasi pada data. dan dapat melakukan hal yang lebih lagi


- Data Frame 2D array
- Series 1D array

## Combining Dataset

```python
df8 = pd.DataFrame({'Fisika':[70,80,90,100],'Matematika':[65,85,75,95],'Biologi':[55,45,65,75]},
                   index=['Andi','Budi','Caca','Dedi'])
df8
```

```python
#merge
countries = pd.DataFrame({'population':[10,20,30],
                                        'continent':['europe','asia','america'],
                                       }, index = ['Holland','Japan','Canada'])
cities = pd.DataFrame({'president_gender':['M','F','M'],
                      'minister':['Yes','No','Yes'],
                      },index=['Holland','Japan','Canada'])
```

```python
countries
```

```python
cities
```

```python
pd.merge(countries, cities, left_index=True,right_index=True)
```
